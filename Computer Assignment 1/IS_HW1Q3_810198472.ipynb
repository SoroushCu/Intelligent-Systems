```python
import numpy as np
import pandas as pd 
import seaborn as sn
import matplotlib.pyplot as plt
import warnings
import os
```

```python
from sklearn import datasets
iris = datasets.load_iris()
data = iris.data[:, :2] #data
label = iris.target #label
```

It is obvious to us that x = data and y = label

```python
x = data
y = label
```

```python
from sklearn.model_selection import train_test_split  
```

```python
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
from sklearn.multiclass import OneVsRestClassifier
```

```python
classifier= OneVsRestClassifier(SVC(kernel = 'linear')).fit(x, y)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred= classifier.predict(x)  
Accuracy = accuracy_score(y, y_pred)
print(f'Accuracy = {Accuracy}')
```

```python
from sklearn.metrics import confusion_matrix  
cm= confusion_matrix(y, y_pred)  
sn.heatmap(cm, annot=True,square = True)
plt.xlabel('predicted value')
plt.ylabel('true value');
```

```python
cf = cm
cf= cf.astype(float)
cf[:,0]=cf[:,0]/sum(cf[:,0])
cf[:,1]=cf[:,1]/sum(cf[:,1])
cf[:,2]=cf[:,2]/sum(cf[:,2])
sn.heatmap(cf, annot=True,square = True)
plt.xlabel('predicted value')
plt.ylabel('true value');
```

```python
from mlxtend.plotting import plot_decision_regions
plt.figure(figsize=(8,8))
plt.title('Different Classification Regions')
plot_decision_regions(x, y, clf=classifier, legend=3)
plt.show()
```
