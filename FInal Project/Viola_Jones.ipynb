# **Viola-Jones Implementation**

```python
import cv2
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
image_path = "/content/drive/MyDrive/db.jpg"
cascade_Path = "/content/drive/MyDrive/haarcascade_frontalface_default.xml"
```

```python
def detect_faces(image_path, cascade_Path):

    faceCascade = cv2.CascadeClassifier(cascade_Path)

    frame = cv2.imread(image_path)
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = faceCascade.detectMultiScale(gray, flags=cv2.CASCADE_SCALE_IMAGE)
    
    return faces, frame
```

```python
faces, frame = detect_faces(image_path, cascade_Path)
```

```python
def show_detected_faces(faces, frame):
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 3)
    plt.imshow(frame)
    plt.show()
```

```python
show_detected_faces(faces, frame)
```
